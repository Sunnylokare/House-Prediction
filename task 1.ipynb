```python
plt.figure(figsize = (20,8))
ax=snb.countplot(x = 'balcony' , data = df)
plt.xticks(rotation = 90)
for p in ax.patches:
  ax.annotate(int(p.get_height()), (p.get_x()+0.25, p.get_height()+1) , va = 'bottom', color='black')
```
